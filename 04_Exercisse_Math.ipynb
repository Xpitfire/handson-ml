```python
def f(t):
    s1 = np.cos(2*np.pi*t)
    e1 = np.exp(-t)
    return np.multiply(s1, e1)
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)
x, y = np.meshgrid(x, y)
r = np.sqrt(x**2 + y**2)
z = f(r)

s = ax.plot_surface(x, y, z, rstride=1, cstride=1, linewidth=0, antialiased=True)
ax.set_zlim3d(-1, 1)
plt.show()
```

## Batch Gradient Descent

```python
eta = 0.1 # learning rate
n_iterations = 1000
m = 100

theta = np.random.randn(2, 1) # random initialization
theta
```

```python
for iteration in range(n_iterations):
    gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y)
    theta = theta - eta * gradients
```

```python
theta
```

```python
# randomly initialization
theta_global = theta = np.random.randn(2, 1)
```

```python
# plot different learning rate results
for index, eta in enumerate([0.02, 0.1, 0.5]):
    print('Learning Rate:', eta)
    theta = theta_global
    # define basic plot
    plt.plot(X, y, "b.")
    plt.axis([0, 2, 0, 15])
    plot_stride = 2
    for iteration in range(n_iterations):
        gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y)
        theta = theta - eta * gradients
        # plot graph line
        if iteration % plot_stride == 0:
            y_predict = X_new_b.dot(theta)
            plt.plot(X_new, y_predict, "r-")
            plot_stride *= 2
    plt.show()
```

### Conclusion
A too high learning rate causes that the gradient does not stailize and jumps mostly away from the minimum. A too small learning rate can cause that the global minimum is never reached, because the adaption takes to long and can be interrupted by the iteration of the epochs.

Also the Batch Gradient Descent does not perform well on huge training sets, because it requires the entire training set for processing at each step. 

This can be improved by the Stohastic Gradient Descent!

## Stohastic Gradient Descent

```python
# with Simulated Annealing approach for the learning rate

n_epochs = 50
t0, t1 = 5, 50 # learning schedule hyperparameters

def learning_schedule(t):
    return t0 / (t + t1)

theta = np.random.randn(2, 1) # random initialization
```

```python
for epoch in range(n_epochs):
    for i in range(m):
        random_index = np.random.randint(m)
        xi = X_b[random_index : random_index+1]
        yi = y[random_index : random_index+1]
        gradients = 2 * xi.T.dot(xi.dot(theta) - yi)
        eta = learning_schedule(epoch * m + 1)
        theta = theta - eta * gradients
```

```python
theta
```

### Conclusion

This algorithm is well suited for large training sets, because it randomly selects values from the training set and over multiple iterations this approach can better approach the global minimum (can also avoid getting stuck in local minima).

But Stohastic Gradient Descent can also become to unstable if the learning rate is too high and never converge towards the global minima. This is the reason why we introduced a Simulated Annealing approach to adapt the learning rate over time and stabilize the gradients towards the end.

### Experiment for plotting factorial dimensionality

```python
import math as math

# combinations
a = np.linspace(1, 100, 100).reshape(100, 1)
d = np.linspace(1, 10, 10)
```

```python
# PolynomialFeatures(degree=d)
# (n + d)! / (d! n!)
def fact_dim(a, d):
    return math.factorial(a + d) / (math.factorial(a) * math.factorial(d))
```

```python
C = np.empty((a.size * d.size), dtype=a.dtype).reshape(a.size, d.size)
C[0::1] = a
```

```python
for i in range(len(C)):
    for j in range(len(C[i])):
        C[i][j] = fact_dim(C[i][j], j + 1)
```
